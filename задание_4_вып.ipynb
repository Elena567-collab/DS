```python
import requests
from bs4 import BeautifulSoup 
import pandas as pd
```

```python
#указываем адрес страницы, с которой будем собирать данные
link = 'https://mybook.ru/sets/12346-top-aprelya-2023/'
#отправка запроса с помощью метода `get()` и запись ответа в переменную result
result = requests.get(link)
```

```python
# проверка возвращаемого кода -- должно быть 200
result.status_code
```

```python
#просмотр текста ответа
result.text
```

```python
# использование конструктура BeautifulSoup для разбора ответа  
soup = BeautifulSoup(result.text, 'lxml')
#print(soup.prettify()) 
```

```python
soup.find('div')
                     
```

```python
# длина всего списка
len(soup.find_all('div'))
```

```python
# поиск нужного тега (здесь это div) с нужным параметром и значением
books = soup.find_all('div', attrs = {'class' : ['e4xwgl-0 iJwsmp']})
```

```python
#определяем длину результата, чтобы проверить, то ли мы нашли
len(books)
```

```python
# поиск нужного тега (здесь это div) с нужным параметром и значением
books[0].find('div', attrs = {'class' : 'dey4wx-1 jVKkXg'}).get_text(strip=True)
```

```python
# поиск нужного тега (здесь это p) с нужным параметром и значением
books[0].find('p', attrs = {'class' : 'lnjchu-1 hhskLb'}).get_text(strip=True)
```

```python
#склеиваем, чтобы получить правильную ссылку
'https://mybook.ru/' + books[0].find('a').get('href')
```

```python
#создаем спискок для хранения результатов -- каждый элемент содержит ровно одну книгу
list_books = list()

#в словать будем помещать информацию ровно об одной книге, а дальше этот словарь будем писать в список  
dict_book = dict.fromkeys(['author', 'title', 'url'], 0)
```

```python
#в цикле обрабатываем найденные записи
for item in books:
    #информацию о каждой книге пишем в словарь
    dict_book['author'] = item.find('div', attrs = {'class':'dey4wx-1 jVKkXg'}).get_text(strip = True)
    dict_book['title'] = item.find('p', attrs = {'class':'lnjchu-1 hhskLb'}).get_text(strip = True)
    dict_book['url'] = '' + item.find('a').get('href')
    #словарь пишем в список
    list_books.append(dict_book.copy())
```

```python
#результирующий датафрейм
pd.DataFrame(list_books)
```
